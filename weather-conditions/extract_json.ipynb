```python
import pandas as pd
```

```python
df = pd.read_csv("./data/conditions_en.csv")
```

```python
df.head()
```

```python
conditions = df[df['address'].str.contains("type")]
```

```python
conditions.head()
```

```python
conditions.to_json("./data/conditions_en.json", index=False)
```

```python
import json
```

```python
conditions_dict = {}
```

```python
codes = conditions["address"].to_numpy()
desc = conditions["Address"].to_numpy()
```

```python
for i in range(len(codes)):
    conditions_dict[codes[i]] = desc[i]
```

```python
conditions_dict
```

```python
conditions_json = json.dumps(conditions_dict)
```

```python
conditions_json
```

```python
with open("conditions_en.json", "a") as f:
    f.write(conditions_json)
```
